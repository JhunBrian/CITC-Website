```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
cs = cv2.imread('images/icons/CS logo.jpg')
cs = cv2.cvtColor(cs, cv2.COLOR_BGR2RGB)
```

```python
cs = cv2.resize(cs, (720, 720))
```

```python
def gaussian_blur(image, kernel_size=(5, 5), sigma_x=0):
    blurred_image = cv2.GaussianBlur(image, kernel_size, sigma_x)
    return blurred_image
```

```python
cs_b = gaussian_blur(cs, kernel_size=(5, 5))
plt.imshow(cs_b)
```

```python
kernel = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])

sharpened = cv2.filter2D(cs_b, -10, kernel)
plt.imshow(sharpened)
```

```python
plt.imsave("CS logo1.png", sharpened)
```
